```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import pandas as pd
```

```python
input="https://www.consumerreports.org/cro/a-to-z-index/products/index.htm"
```

```python
us_ag=UserAgent()
```

```python
page=requests.get(input,headers={'user_agent':us_ag.chrome}).text
```

```python
soup=BeautifulSoup(page,'lxml')
soup
```

```python
soup.find_all('a',attrs={'class':'products-a-z__results__item'})
```

```python
for tag in soup.find_all('a',attrs={'class':'products-a-z__results__item'}):
    print(tag.string)
```

```python
product_1=[tag.string for tag in soup.find_all('a',attrs={'class':'products-a-z__results__item'})]
product_1
```

```python
product_2=[product_1[i].replace('\n','') for i in range(len(product_1))]
```

```python
for tag in soup.find_all('a',attrs={'class':'products-a-z__results__item'}):
    print(tag['href'])
```

```python
href_1=[tag['href'] for tag in soup.find_all('a',attrs={'class':'products-a-z__results__item'})]
```

```python
href_2=['https://www.consumerreports.org'+href_1[i] for i in range(len(href_1))]
```

```python
df=pd.DataFrame()
df['product']=product_2
df['href']=href_2
```

```python
df['href'][0]
```

```python
df.to_csv('product.csv')
```
